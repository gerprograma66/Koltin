<details>
<summary style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 0.9rem; font-weight: 500; cursor: pointer; color: var(--jp-ui-font-color2, #bbbbbb); background-color: var(--jp-cell-editor-active-background, #2a2a2a); display: inline-block; padding: 6px 14px; border-radius: 20px; margin-bottom: 16px; transition: background 0.2s; user-select: none;">📘 Ver / ocultar contenido</summary>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Una clase es el <strong style="color: #bb86fc;">molde o plantilla</strong> a partir del cual se crean objetos. Define qu&eacute; datos tendr&aacute; y qu&eacute; comportamiento tendr&aacute; cada objeto creado a partir de ella. En Kotlin se declaran con la palabra clave <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">class</code>.</p>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 0; color: var(--jp-ui-font-color2, #cccccc);">A diferencia de Java, Kotlin hace las clases mucho m&aacute;s <strong style="color: #03dac6;">concisas</strong>. Lo que en Java requer&iacute;a decenas de l&iacute;neas (constructor, getters, setters, etc.), en Kotlin puede expresarse en una sola l&iacute;nea.</p>
</details>

<!-- Details colapsable -->
<details>
<summary style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 0.9rem; font-weight: 500; cursor: pointer; color: var(--jp-ui-font-color2, #bbbbbb); background-color: var(--jp-cell-editor-active-background, #2a2a2a); display: inline-block; padding: 6px 14px; border-radius: 20px; margin-bottom: 16px; transition: background 0.2s; user-select: none;">📘 Ver / ocultar contenido</summary>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Las propiedades son los <strong style="color: #bb86fc;">datos que pertenecen a una clase</strong>. En Kotlin, una propiedad no es simplemente un campo como en Java, sino que internamente incluye:</p>
<ul style="margin: 8px 0 16px 0; padding-left: 24px; font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; color: var(--jp-ui-font-color2, #cccccc);">
<li style="margin: 8px 0;">Un <strong style="color: #03dac6;">campo de respaldo</strong> que almacena el valor.</li>
<li style="margin: 8px 0;">Un <strong style="color: #03dac6;">getter</strong> para leer el valor.</li>
<li style="margin: 8px 0;">Un <strong style="color: #03dac6;">setter</strong> para modificarlo (solo si es <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">var</code>).</li>
</ul>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Kotlin genera todo esto autom&aacute;ticamente, pero tambi&eacute;n te permite personalizarlo si lo necesitas.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);"><code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">val</code> vs <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">var</code></h3>
<ul style="margin: 8px 0 0 0; padding-left: 24px; font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; color: var(--jp-ui-font-color2, #cccccc);">
<li style="margin: 8px 0;"><code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">val</code>: La propiedad es de <strong style="color: #bb86fc;">solo lectura</strong>. Una vez asignada no puede cambiar. Genera solo getter.</li>
<li style="margin: 8px 0;"><code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">var</code>: La propiedad es <strong style="color: #03dac6;">mutable</strong>. Puede cambiar en cualquier momento. Genera getter y setter.</li>
</ul>
</details>

<details>
<summary style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 0.9rem; font-weight: 500; cursor: pointer; color: var(--jp-ui-font-color2, #bbbbbb); background-color: var(--jp-cell-editor-active-background, #2a2a2a); display: inline-block; padding: 6px 14px; border-radius: 20px; margin-bottom: 16px; transition: background 0.2s; user-select: none;">📘 Ver / ocultar contenido</summary>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 0; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">Constructor primario</h3>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">En Kotlin, el constructor principal se define directamente en la cabecera de la clase. Es la forma m&aacute;s <strong style="color: #03dac6;">concisa</strong> y com&uacute;n de declarar una clase. Los par&aacute;metros del constructor primario pueden convertirse directamente en propiedades usando <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">val</code> o <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">var</code>.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">Constructor secundario</h3>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 0; color: var(--jp-ui-font-color2, #cccccc);">Cuando necesitas m&uacute;ltiples formas de crear un objeto, puedes definir constructores secundarios con la palabra clave <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">constructor</code>. Sin embargo, en Kotlin es m&aacute;s com&uacute;n y recomendado usar <strong style="color: #bb86fc;">par&aacute;metros por defecto</strong> en el constructor primario en lugar de constructores secundarios.</p>
</details>

Inicializadores (<code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">init</code>)</h2>
<!-- Details colapsable -->
<details>
<summary style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 0.9rem; font-weight: 500; cursor: pointer; color: var(--jp-ui-font-color2, #bbbbbb); background-color: var(--jp-cell-editor-active-background, #2a2a2a); display: inline-block; padding: 6px 14px; border-radius: 20px; margin-bottom: 16px; transition: background 0.2s; user-select: none;">📘 Ver / ocultar contenido</summary>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">El bloque <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">init</code> es el lugar donde puedes ejecutar <strong style="color: #03dac6;">l&oacute;gica de inicializaci&oacute;n</strong> cuando se crea un objeto. Se ejecuta inmediatamente despu&eacute;s del constructor primario, en el orden en que aparece en el cuerpo de la clase.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">&iquest;Por qu&eacute; existe si ya hay un constructor?</h3>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">El constructor primario en Kotlin no puede contener c&oacute;digo directamente, solo declarar par&aacute;metros. El bloque <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">init</code> llena ese vac&iacute;o, permiti&eacute;ndote <strong style="color: #bb86fc;">validar datos, transformar valores o ejecutar cualquier l&oacute;gica</strong> necesaria al momento de crear el objeto.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">M&uacute;ltiples bloques <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">init</code></h3>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 0; color: var(--jp-ui-font-color2, #cccccc);">Puedes tener varios bloques <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">init</code> en una misma clase. Se ejecutan en el <strong style="color: #03dac6;">orden en que aparecen</strong>, intercalados con la inicializaci&oacute;n de propiedades. Esto es &uacute;til para separar l&oacute;gica de inicializaci&oacute;n en pasos claros.</p>
</details>

```kotlin
// Tenemnos una tienda, con multiples clientes, cada cliente tiene un saldo, un nombre y una direccion
// para cada cliente puede efectuar un abono que reduce el saldo, una vez que el saldo llega a cero no puedo
// hacer mas abonos, si la cantidad que estoy abonando al saldo tiene un residuo entonces debo dar cambio
// Tambien me debe mostrar la informacion del cliente cuando se solicite

        //Constructor Primario
class Cliente(var saldo : Float, val nombre: String, val direccion : String ){
var esEstudiante = false

        init {
                this.mostrarInfoCliente()
            println("Init 1")
        }
            init {
                println("Init 2")
            }
            init {
                println("Init 3")
            }


    //Constructor Secundario para registrar clientes que son estudiantes
    constructor(saldo : Float, nombre: String, direccion: String, esEstudiante: Boolean) : this(saldo, nombre, direccion)
    {
       this.esEstudiante = esEstudiante
        val estudiante = if (esEstudiante) "Sí" else "No"

        println("Es estudiante: $estudiante")

    }

    fun mostrarInfoCliente(){
        println("Nombre: $nombre")
        println("Direccion: $direccion")
        println("Saldo Actual: $saldo")
    }

    fun efectuarAbono(cantidadAbonada: Float) : Float{
        var cambio =  0.0f
        if(saldo == 0.0f){
            println("No es posible efectuar abono, saldo actual: $saldo")
        }else{
                            //100   -    30  = 70
                            //100   -    200 = -100
            val saldoActual = saldo - cantidadAbonada

            if(saldoActual < 0.0f) {
            saldo = 0.0f
            cambio = saldoActual * -1
            }else{
                saldo = saldoActual
            }
        }

        println("El nuevo saldo es: $$saldo, Cambio a entregar: $$cambio")
        return cambio
    }
}
```
