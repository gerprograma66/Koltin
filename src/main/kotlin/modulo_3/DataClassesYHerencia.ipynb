<details>
<summary style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 0.9rem; font-weight: 500; cursor: pointer; color: var(--jp-ui-font-color2, #bbbbbb); background-color: var(--jp-cell-editor-active-background, #2a2a2a); display: inline-block; padding: 6px 14px; border-radius: 20px; margin-bottom: 16px; transition: background 0.2s; user-select: none;">📘 Ver / ocultar contenido</summary>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Una <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">data class</code> es una clase dise&ntilde;ada espec&iacute;ficamente para <strong style="color: #03dac6;">almacenar y transportar datos</strong>. En Java, crear una clase as&iacute; requer&iacute;a escribir manualmente constructores, getters, <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">equals()</code>, <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">hashCode()</code> y <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">toString()</code>. En Kotlin, simplemente agregando la palabra clave <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">data</code>, el compilador genera todo eso autom&aacute;ticamente.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">&iquest;Qu&eacute; genera autom&aacute;ticamente?</h3>
<ul style="margin: 8px 0 16px 0; padding-left: 24px; font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; color: var(--jp-ui-font-color2, #cccccc);">
<li style="margin: 8px 0;"><strong><code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">toString()</code></strong>: Representaci&oacute;n legible del objeto con todos sus valores.</li>
<li style="margin: 8px 0;"><strong><code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">equals()</code></strong>: Compara dos objetos por el <strong style="color: #bb86fc;">valor de sus propiedades</strong>, no por referencia en memoria.</li>
<li style="margin: 8px 0;"><strong><code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">hashCode()</code></strong>: Consistente con <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">equals()</code>, necesario para usar objetos en colecciones como <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">HashMap</code> o <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">HashSet</code>.</li>
<li style="margin: 8px 0;"><strong><code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">copy()</code></strong>: Crea una copia del objeto permitiendo modificar solo algunas propiedades.</li>
</ul>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">Restricciones importantes</h3>
<ul style="margin: 8px 0 0 0; padding-left: 24px; font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; color: var(--jp-ui-font-color2, #cccccc);">
<li style="margin: 8px 0;">Las propiedades del constructor primario deben ser <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">val</code> o <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">var</code>.</li>
<li style="margin: 8px 0;">No pueden ser <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">abstract</code>, <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">open</code>, <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">sealed</code> ni <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">inner</code>.</li>
</ul>

<!-- Details colapsable -->
<details>
<summary style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 0.9rem; font-weight: 500; cursor: pointer; color: var(--jp-ui-font-color2, #bbbbbb); background-color: var(--jp-cell-editor-active-background, #2a2a2a); display: inline-block; padding: 6px 14px; border-radius: 20px; margin-bottom: 16px; transition: background 0.2s; user-select: none;">📘 Ver / ocultar contenido</summary>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">La herencia permite que una clase <strong style="color: #bb86fc;">adquiera las propiedades y m&eacute;todos de otra</strong>, evitando duplicar c&oacute;digo y estableciendo una relaci&oacute;n "es un tipo de".</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);"><code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">open</code> &mdash; Habilitando la herencia</h3>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">En Kotlin, <strong style="color: #03dac6;">todas las clases son <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">final</code> por defecto</strong>, lo que significa que no pueden ser heredadas a menos que se marquen expl&iacute;citamente con <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">open</code>. Esto es una decisi&oacute;n de dise&ntilde;o intencional: obliga al programador a pensar conscientemente si una clase est&aacute; dise&ntilde;ada para ser extendida.</p>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Lo mismo aplica para los m&eacute;todos: si quieres que una subclase pueda sobreescribirlo, debe marcarse con <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">open</code>. Para sobreescribirlo en la subclase se usa <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">override</code>.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">Llamando al padre con <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">super</code></h3>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 0; color: var(--jp-ui-font-color2, #cccccc);">Cuando sobreescribes un m&eacute;todo pero tambi&eacute;n quieres ejecutar la l&oacute;gica del padre, usas <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">super</code>. Esto es &uacute;til cuando la subclase <strong style="color: #bb86fc;">extiende el comportamiento</strong> en lugar de reemplazarlo completamente.</p>
</details>

Clases Abstractas (<code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">abstract</code>)</h2>

<details>
<summary style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 0.9rem; font-weight: 500; cursor: pointer; color: var(--jp-ui-font-color2, #bbbbbb); background-color: var(--jp-cell-editor-active-background, #2a2a2a); display: inline-block; padding: 6px 14px; border-radius: 20px; margin-bottom: 16px; transition: background 0.2s; user-select: none;">📘 Ver / ocultar contenido</summary>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Una clase abstracta es una clase que <strong style="color: #ffb86c;">no puede instanciarse directamente</strong>. Existe &uacute;nicamente para ser heredada. Puede tener m&eacute;todos con implementaci&oacute;n y m&eacute;todos abstractos (sin implementaci&oacute;n) que las subclases est&aacute;n <strong style="color: #03dac6;">obligadas a implementar</strong>.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">&iquest;Cu&aacute;ndo usar <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">abstract</code> en lugar de <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">open</code>?</h3>
<ul style="margin: 8px 0 0 0; padding-left: 24px; font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; color: var(--jp-ui-font-color2, #cccccc);">
<li style="margin: 8px 0;">Usa <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">open</code> cuando la clase tiene sentido por s&iacute; sola pero tambi&eacute;n puede extenderse.</li>
<li style="margin: 8px 0;">Usa <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">abstract</code> cuando la clase es incompleta por dise&ntilde;o y <strong style="color: #bb86fc;">no tiene sentido crear objetos de ella directamente</strong>.</li>
</ul>
