<h3 style="margin: 0; font-size: 1.8rem; font-weight: 600; color: #ccc5f0; line-height: 1.5;">Una <strong style="color: #e0d4ff;">condici&oacute;n</strong> es una pregunta que solo puede tener dos respuestas: <strong style="color: #a5d6ff;">verdadero (<code style="background: rgba(127, 82, 255, 0.25); padding: 5px 12px; border-radius: 10px; font-family: monospace; color: #a5d6ff; font-size: 1.6rem;">true</code>)</strong> o <strong style="color: #ffb4a2;">falso (<code style="background: rgba(127, 82, 255, 0.25); padding: 5px 12px; border-radius: 10px; font-family: monospace; color: #ffb4a2; font-size: 1.6rem;">false</code>)</strong>.</h3>

<!-- Lista de características (nuevo contenido con estilo uniforme) -->
<ul style="margin: 28px 0; padding-left: 40px; color: #b8b0e0; font-size: 1.5rem; line-height: 1.6;">
<li style="margin: 16px 0; list-style-type: '✨ ';">Eval&uacute;a &uacute;nicamente <strong style="color: #e0d4ff;">expresiones booleanas</strong> (<code style="background: rgba(127, 82, 255, 0.25); padding: 5px 12px; border-radius: 10px; font-family: monospace; color: #b39ddb; font-size: 1.4rem;">true</code> / <code style="background: rgba(127, 82, 255, 0.25); padding: 5px 12px; border-radius: 10px; font-family: monospace; color: #b39ddb; font-size: 1.4rem;">false</code>), no acepta enteros u otros tipos como condici&oacute;n.</li>
<li style="margin: 16px 0; list-style-type: '📌 ';">El bloque <code style="background: rgba(127, 82, 255, 0.25); padding: 5px 12px; border-radius: 10px; font-family: monospace; color: #b39ddb; font-size: 1.4rem;">else</code> es <strong style="color: #ffb4a2;">opcional</strong> cuando se usa como sentencia, pero <strong style="color: #a5d6ff;">obligatorio</strong> cuando se usa como expresi&oacute;n.</li>
<li style="margin: 16px 0; list-style-type: '🎯 ';">Cuando se usa como expresi&oacute;n, retorna el <strong style="color: #e0d4ff;">valor de la &uacute;ltima l&iacute;nea</strong> del bloque ejecutado.</li>
<li style="margin: 16px 0; list-style-type: '🔗 ';">Puede encadenarse con <code style="background: rgba(127, 82, 255, 0.25); padding: 5px 12px; border-radius: 10px; font-family: monospace; color: #b39ddb; font-size: 1.4rem;">else if</code> para evaluar <strong style="color: #e0d4ff;">m&uacute;ltiples condiciones</strong> de forma secuencial.</li>
<li style="margin: 16px 0; list-style-type: '⬇️ ';">Las condiciones se eval&uacute;an <strong style="color: #e0d4ff;">de arriba hacia abajo</strong> y se detiene en la primera que sea verdadera.</li>
<li style="margin: 16px 0; list-style-type: '🔄 ';"><strong style="color: #ffd966;">Reemplaza al operador ternario</strong> <code style="background: rgba(127, 82, 255, 0.25); padding: 5px 12px; border-radius: 10px; font-family: monospace; color: #b39ddb; font-size: 1.4rem;">? :</code> de Java/C, ya que cumple la misma funci&oacute;n como expresi&oacute;n.</li>
</ul>
