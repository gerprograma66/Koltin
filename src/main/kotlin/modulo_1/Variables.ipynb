Se definen utilizando la palabra reservada <code style="background: rgba(124, 77, 255, 0.25); color: #c792ea; padding: 6px 14px; border-radius: 12px; font-weight: bold; font-family: monospace; font-size: 1.5rem;">var</code> .</p>

<li style="margin: 12px 0; line-height: 1.4;"><strong style="color: #03dac6;">Inmutabilidad</strong>: Facilita el razonamiento sobre el c&oacute;digo y el multithreading</li>
