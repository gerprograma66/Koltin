<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Una lambda es una <strong style="color: #bb86fc;">funci&oacute;n an&oacute;nima</strong>, es decir, una funci&oacute;n que no tiene nombre y puede tratarse como un valor m&aacute;s del programa. Esto significa que puedes guardarla en una variable, pasarla como argumento a otra funci&oacute;n o devolverla como resultado.</p>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Su sintaxis usa llaves <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">{}</code> y dentro de ellas se definen los par&aacute;metros, seguidos de <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">-&gt;</code> y luego el cuerpo de la funci&oacute;n. Una caracter&iacute;stica importante es que <strong style="color: #03dac6;">no necesitas escribir <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">return</code></strong>, ya que el &uacute;ltimo valor evaluado dentro de la lambda es autom&aacute;ticamente su resultado.</p>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 0; color: var(--jp-ui-font-color2, #cccccc);">El tipo de una lambda se expresa como <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">(TipoParam1, TipoParam2) -&gt; TipoRetorno</code>. Si no retorna nada &uacute;til, el tipo de retorno es <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">Unit</code>.</p>

El par&aacute;metro <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">it</code></h2>

<details>
<summary style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 0.9rem; font-weight: 500; cursor: pointer; color: var(--jp-ui-font-color2, #bbbbbb); background-color: var(--jp-cell-editor-active-background, #2a2a2a); display: inline-block; padding: 6px 14px; border-radius: 20px; margin-bottom: 16px; transition: background 0.2s; user-select: none;">📘 Ver / ocultar contenido</summary>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Cuando una lambda recibe <strong style="color: #bb86fc;">un &uacute;nico par&aacute;metro</strong>, Kotlin ofrece un atajo: puedes omitir completamente la declaraci&oacute;n del par&aacute;metro y la flecha <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">-&gt;</code>, y referenciarlo autom&aacute;ticamente usando la palabra reservada <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">it</code>.</p>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Esto existe porque en la pr&aacute;ctica, muchas lambdas que se usan con colecciones (<code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">map</code>, <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">filter</code>, <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">forEach</code>, etc.) solo reciben un elemento a la vez, por lo que nombrar ese par&aacute;metro expl&iacute;citamente ser&iacute;a repetitivo e innecesario.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">Cu&aacute;ndo NO usar <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">it</code></h3>
<ul style="margin: 8px 0 16px 0; padding-left: 24px; font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; color: var(--jp-ui-font-color2, #cccccc);">
<li style="margin: 8px 0;">Cuando la lambda tiene <strong style="color: #ffb86c;">m&aacute;s de un par&aacute;metro</strong>, <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">it</code> no est&aacute; disponible.</li>
<li style="margin: 8px 0;">Cuando el c&oacute;digo se vuelve confuso por tener lambdas anidadas, cada una con su propio <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">it</code>. En ese caso es mejor nombrar los par&aacute;metros expl&iacute;citamente para mayor claridad.</li>
</ul>

<!-- Details colapsable -->
<details>
<summary style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 0.9rem; font-weight: 500; cursor: pointer; color: var(--jp-ui-font-color2, #bbbbbb); background-color: var(--jp-cell-editor-active-background, #2a2a2a); display: inline-block; padding: 6px 14px; border-radius: 20px; margin-bottom: 16px; transition: background 0.2s; user-select: none;">📘 Ver / ocultar contenido</summary>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Una funci&oacute;n de orden superior es aquella que <strong style="color: #bb86fc;">recibe una o m&aacute;s funciones como par&aacute;metros</strong>, o que <strong style="color: #bb86fc;">devuelve una funci&oacute;n</strong> como resultado. Esto es posible porque en Kotlin las funciones son <strong style="color: #03dac6;">ciudadanos de primera clase</strong>, lo que significa que pueden usarse como cualquier otro tipo de dato.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">&iquest;Por qu&eacute; son importantes?</h3>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Permiten escribir c&oacute;digo mucho m&aacute;s <strong style="color: #03dac6;">gen&eacute;rico y reutilizable</strong>. En lugar de duplicar l&oacute;gica que solo var&iacute;a en un peque&ntilde;o comportamiento, puedes extraer ese comportamiento variable como una funci&oacute;n y pasarla como par&aacute;metro.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">Trailing Lambda</h3>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Existe una convenci&oacute;n muy importante en Kotlin: si el <strong style="color: #bb86fc;">&uacute;ltimo par&aacute;metro</strong> de una funci&oacute;n es una funci&oacute;n, puedes sacar la lambda <strong style="color: #03dac6;">fuera de los par&eacute;ntesis</strong>. Esto hace que el c&oacute;digo se lea de forma m&aacute;s natural, casi como si fuera una estructura del lenguaje.</p>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 0; color: var(--jp-ui-font-color2, #cccccc);">Esta convenci&oacute;n es la raz&oacute;n por la que funciones como <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">apply</code>, <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">run</code>, <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">let</code>, <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">also</code> y <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">with</code> en Kotlin se ven tan naturales al usarse.</p>
