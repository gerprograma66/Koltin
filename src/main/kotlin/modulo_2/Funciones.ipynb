<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Una funci&oacute;n en Kotlin es un bloque de c&oacute;digo reutilizable que realiza una tarea espec&iacute;fica. Se define con la palabra clave <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">fun</code>, seguida del nombre, los par&aacute;metros entre par&eacute;ntesis y el tipo de retorno.</p>

El compilador infiere autom&aacute;ticamente el tipo de retorno en estos casos.</p>

<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Son valores que se asignan a un par&aacute;metro directamente en la <strong>definici&oacute;n de la funci&oacute;n</strong>. <h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">&iquest;Por qu&eacute; son &uacute;tiles?</h3>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">En Java, cuando quer&iacute;as tener una funci&oacute;n con comportamientos opcionales, ten&iacute;as que crear <strong>m&uacute;ltiples versiones de la misma funci&oacute;n</strong> (sobrecarga). En Kotlin, con una sola funci&oacute;n puedes cubrir todos esos casos.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">Comportamiento importante</h3>
<ul style="margin: 8px 0 16px 0; padding-left: 24px; font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; color: var(--jp-ui-font-color2, #cccccc);">
<li style="margin: 8px 0;">Puedes tener <strong>varios par&aacute;metros con valor por defecto</strong> en la misma funci&oacute;n.</li>
<li style="margin: 8px 0;">Los par&aacute;metros con valor por defecto suelen colocarse <strong>al final</strong> de la lista de par&aacute;metros, aunque no es obligatorio.</li>
<li style="margin: 8px 0;">Si un par&aacute;metro sin valor por defecto viene <strong>despu&eacute;s</strong> de uno con valor por defecto, es obligatorio usar <strong>par&aacute;metros nombrados</strong> para evitar ambig&uuml;edad.</li>

<details>
<summary style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 0.9rem; font-weight: 500; cursor: pointer; color: var(--jp-ui-font-color2, #bbbbbb); background-color: var(--jp-cell-editor-active-background, #2a2a2a); display: inline-block; padding: 6px 14px; border-radius: 20px; margin-bottom: 16px; transition: background 0.2s; user-select: none;">📘 Ver / ocultar contenido</summary>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 0; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">&iquest;Qu&eacute; son?</h3>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Es la capacidad de especificar <strong style="color: #bb86fc;">el nombre del par&aacute;metro</strong> al momento de llamar una funci&oacute;n, en lugar de depender &uacute;nicamente del <strong style="color: #03dac6;">orden</strong> en que fueron declarados.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">&iquest;Por qu&eacute; son &uacute;tiles?</h3>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Cuando una funci&oacute;n tiene muchos par&aacute;metros, especialmente del <strong style="color: #ffb86c;">mismo tipo</strong> (como varios <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">String</code> o varios <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">Boolean</code>), es muy f&aacute;cil confundirse y pasar los valores en el <strong style="color: #ffb86c;">orden incorrecto</strong>. Los <strong style="color: #bb86fc;">par&aacute;metros nombrados</strong> eliminan esa ambig&uuml;edad.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">Comportamiento importante</h3>
<ul style="margin: 8px 0 16px 0; padding-left: 24px; font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; color: var(--jp-ui-font-color2, #cccccc);">
<li style="margin: 8px 0;">Puedes <strong style="color: #03dac6;">mezclar</strong> argumentos posicionales y nombrados, pero una vez que usas un argumento nombrado, los siguientes tambi&eacute;n deben ser nombrados.</li>
<li style="margin: 8px 0;">Mejoran enormemente la <strong style="color: #bb86fc;">legibilidad del c&oacute;digo</strong>, ya que quien lee la llamada a la funci&oacute;n entiende inmediatamente qu&eacute; representa cada valor.</li>
<li style="margin: 8px 0;">Funcionan muy bien en combinaci&oacute;n con los <strong style="color: #03dac6;">par&aacute;metros por defecto</strong>, ya que puedes omitir solo los par&aacute;metros que no necesitas cambiar, sin importar su posici&oacute;n.</li>
</ul>
