<details>
<summary style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 0.9rem; font-weight: 500; cursor: pointer; color: var(--jp-ui-font-color2, #bbbbbb); background-color: var(--jp-cell-editor-active-background, #2a2a2a); display: inline-block; padding: 6px 14px; border-radius: 20px; margin-bottom: 16px; transition: background 0.2s; user-select: none;">📘 Ver / ocultar contenido</summary>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 0; color: var(--jp-ui-font-color2, #cccccc);">En Java y muchos otros lenguajes, cualquier variable puede ser <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">null</code> en cualquier momento, lo que lleva al famoso <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">NullPointerException</code> (NPE). Kotlin resuelve esto desde el <strong style="color: #bb86fc;">sistema de tipos</strong>: por defecto, ninguna variable puede ser <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">null</code> a menos que lo indiques expl&iacute;citamente. Esto convierte los errores de null en <strong style="color: #03dac6;">errores de compilaci&oacute;n</strong> en lugar de errores en tiempo de ejecuci&oacute;n.</p>
</details>

Tipos Nullable (<code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">?</code>)</h2>

<details>
<summary style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 0.9rem; font-weight: 500; cursor: pointer; color: var(--jp-ui-font-color2, #bbbbbb); background-color: var(--jp-cell-editor-active-background, #2a2a2a); display: inline-block; padding: 6px 14px; border-radius: 20px; margin-bottom: 16px; transition: background 0.2s; user-select: none;">📘 Ver / ocultar contenido</summary>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">En Kotlin, los tipos se dividen en dos categor&iacute;as:</p>
<ul style="margin: 8px 0 16px 0; padding-left: 24px; font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; color: var(--jp-ui-font-color2, #cccccc);">
<li style="margin: 8px 0;"><strong style="color: #03dac6;">No nullable:</strong> No pueden contener <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">null</code> bajo ninguna circunstancia. El compilador lo garantiza.</li>
<li style="margin: 8px 0;"><strong style="color: #bb86fc;">Nullable:</strong> Se declaran agregando <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">?</code> al tipo. Indican que la variable <strong style="color: #ffb86c;">puede o no</strong> tener un valor.</li>
</ul>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 0; color: var(--jp-ui-font-color2, #cccccc);">Esta distinci&oacute;n obliga al programador a <strong style="color: #03dac6;">manejar expl&iacute;citamente</strong> la posibilidad de null antes de usar una variable nullable, eliminando los NPE accidentales.</p>
</details>

<details>
<summary style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 0.9rem; font-weight: 500; cursor: pointer; color: var(--jp-ui-font-color2, #bbbbbb); background-color: var(--jp-cell-editor-active-background, #2a2a2a); display: inline-block; padding: 6px 14px; border-radius: 20px; margin-bottom: 16px; transition: background 0.2s; user-select: none;">📘 Ver / ocultar contenido</summary>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Cuando tienes una variable nullable y quieres acceder a sus propiedades o m&eacute;todos, no puedes hacerlo directamente porque el compilador no lo permite. El operador <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">?.</code> resuelve esto: <strong style="color: #03dac6;">solo ejecuta la operaci&oacute;n si el valor no es null</strong>. Si es null, simplemente retorna <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">null</code> sin lanzar ninguna excepci&oacute;n.</p>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 0; color: var(--jp-ui-font-color2, #cccccc);">Esto permite encadenar m&uacute;ltiples operaciones de forma segura. Si en alg&uacute;n punto de la cadena hay un <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">null</code>, toda la cadena retorna <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">null</code> sin explotar.</p>

<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">El operador <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">?:</code> es el complemento natural de <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">?.</code>. Permite definir un <strong style="color: #03dac6;">valor por defecto</strong> que se usar&aacute; cuando la expresi&oacute;n de la izquierda sea <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">null</code>. Su nombre viene de que si lo giras 90&deg;, se parece al peinado de Elvis Presley 🕺.</p>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 0; color: var(--jp-ui-font-color2, #cccccc);">Es extremadamente &uacute;til para evitar valores nulos en cadena y garantizar que siempre tengas un <strong style="color: #bb86fc;">valor con el que trabajar</strong>.</p>

Si en tiempo de ejecuci&oacute;n el valor <strong style="color: #ffb86c;">s&iacute; es null</strong>, lanza una <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">NullPointerException</code>.</p>
<h3 style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1.2rem; font-weight: 500; margin-top: 16px; margin-bottom: 8px; color: var(--jp-ui-font-color1, #e0e0e0);">&iquest;Cu&aacute;ndo usarlo?</h3>
<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);">Pr&aacute;cticamente <strong style="color: #ffb86c;">nunca</strong>, o solo en casos muy espec&iacute;ficos donde tienes <strong style="color: #03dac6;">certeza absoluta</strong> de que el valor no puede ser null en ese punto, pero el compilador no puede inferirlo por s&iacute; solo.</p>

<p style="font-family: 'Segoe UI', system-ui, sans-serif; font-size: 1rem; line-height: 1.5; margin-bottom: 16px; color: var(--jp-ui-font-color2, #cccccc);"><code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">let</code> es una funci&oacute;n de alcance que ejecuta un bloque de c&oacute;digo <strong style="color: #03dac6;">solo si el valor no es null</strong>, usando <code style="background-color: rgba(124, 77, 255, 0.2); padding: 2px 6px; border-radius: 6px; font-family: 'JetBrains Mono', 'Fira Code', monospace; color: #c792ea;">?.let</code>.
